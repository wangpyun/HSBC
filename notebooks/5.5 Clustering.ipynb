```python
import numpy as np
from sklearn.manifold import TSNE
import plotly.graph_objs as go
from plotly.subplots import make_subplots


emb_file = '/Users/Pengyun_Wang/Projects/HSBC/data/gte-embeddings.npy'
# response_file = '/Users/Pengyun_Wang/Projects/HSBC/data/openai_responses.json'
# with open(response_file, 'r') as f:
#     response_data = json.load(f)
embeddings = np.load(emb_file)

# Perform t-SNE dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(embeddings)

# Create scatter plot
scatter = go.Scatter(
    x=embeddings_2d[:, 0],
    y=embeddings_2d[:, 1],
    mode='markers',
    marker=dict(color='blue', size=5),
    name='Embeddings'
)

# Create plotly figure
fig = make_subplots(rows=1, cols=1)
fig.add_trace(scatter)

# Update layout
fig.update_layout(
    title='t-SNE Visualization of Embeddings',
    xaxis=dict(title='t-SNE Component 1'),
    yaxis=dict(title='t-SNE Component 2'),
    showlegend=True
)

# Show interactive plot
fig.show()
```
